### Inference 

#### The missing values and the data correction has been performed and imputed the missing values with mean n mode value for both training and testing data.

```python
df_cat_train = df_train.select_dtypes(include = ['object'])
df_num_train = df_train.select_dtypes(include = ['float64','int64'])
```

```python
df_cat_test = df_test.select_dtypes(include = ['object'])
df_num_test = df_test.select_dtypes(include = ['float64','int64'])
```

```python
df_num_train.info()
```

```python
df_test.select_dtypes(include = ['float64','int64']).info()
```

```python
for x in df_num_train:
#     print("----before-----",df_num_train[x].dtype)
    if(x != 'Attrition'):
        if df_num_train[x].dtype != df_test[x].dtype:
             df_num_train[x] = df_num_train[x].astype(np.int64)
#             print(df_num_train[x].dtype)    
    else:
         df_num_train[x] = df_num_train[x].astype(np.int64)
    
```

```python
df_num_train.info()
```

```python
df_cat_train = pd.get_dummies(df_cat_train, drop_first = True)
df_cat_train.head()
```

```python
df_cat_test = pd.get_dummies(df_cat_test, drop_first = True)
df_cat_test.head()
```

```python
df_train = pd.concat([df_cat_train,df_num_train],axis=1)
df_train.head()
```

```python
df_test = pd.concat([df_cat_test,df_num_test],axis=1)
df_test.head()
```

```python
df_train_copy = df_train.drop(columns=['EmployeeID'], axis=1)
df_train_copy.info()
```

```python
test_file_ids = df_test['EmployeeID'] ## necessary step
df_test_copy = df_test.drop(columns=['EmployeeID'], axis=1)
df_test_copy.info()
```

```python
df_train_copy.shape
```

```python
x = df_train_copy.drop('Attrition', axis=1)

y = df_train_copy['Attrition']
```

x = pd.concat([x, df_test_copy],axis=0, ignore_index=True)
x.shape

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.252 , random_state=42,stratify=y)
x_train.shape
```

### Logistic Regression

```python
grid={'penalty':['l1','l2','none'],
      'solver':['lbfgs', 'liblinear'],
      'tol':[0.0001,0.000001]}
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=100000,n_jobs=2)
```

```python
grid_search = GridSearchCV(estimator = model, param_grid = grid, cv = 3,n_jobs=-1,scoring='f1')
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
ytrain_predict = best_model.predict(x_train)
ytest_predict = best_model.predict(x_test)
ytrain_predict
```

```python
ytest_predict
```

```python
Logistic_regression = grid_search.score(x_test, y_test)
Logistic_regression
```
